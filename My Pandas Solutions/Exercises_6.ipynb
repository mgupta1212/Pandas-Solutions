### Challenge: first combine these sales together into a single dataframe, then compute how much money consumers spent on each book in each currency.

```python
newdf1 = sales1.rename(columns={'Book title': 'title', 'Number sold': 'copies_sold', 'Sales price': 'sale_price', 'Royalty paid': 'royalty',})
newdf1['currency'] = 'USD'
newdf1
```

```python
s2 = sales2.copy()
titles = sales2['Title']
titles = titles.where(titles.str.endswith(')')).str.split().str[-1].str.strip('()')
s2['currency'] = titles.fillna(method='bfill')
s2
```

```python
s2 = s2[s2['List price'].notnull()]
s2
```

```python
newdf2 = s2.rename(columns={'Title': 'title', 'Units sold': 'copies_sold', 'List price': 'sale_price', 'Royalty': 'royalty',})
newdf2.head()
```

```python
newdf = pd.concat([newdf1,newdf2])
newdf
```

```python
newdfcopy = newdf.assign(total = newdf.copies_sold * newdf.sale_price)
newdfcopy = newdfcopy.assign(new_title = newdfcopy['title'].str.lower().str.replace("’", "'"))
newdfcopy.groupby(['new_title','currency'])[['total']].sum()
```
